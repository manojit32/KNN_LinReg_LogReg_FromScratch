### Using Scikit-learn Naive Bayes :-

```python
import pandas as pd
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
df=pd.read_csv("LoanDataset/data.csv")
X=df.iloc[:,[1,2,3,5,6,7,8,10,11,12,13]]
y=df.iloc[:,9]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
```

### Observations :
Our own algorithm works better than the algorithm that is given by Scikit Learn, because in Scikit-Learn, we can either use GaussianNB which is good for numerical features, or we can use MultinomialNB, which is used only for Categorical Features, we cannot mix and match those two together using Scikit-Learn in case we have mixed data like this. Using own algorithm, we can use Gaussian and Multinomial both in this data, so the accuracy is higher in case of our model.

## Part 4.3 : Plot a graph explaining the co-relation between threshold value vs precision and recall. Report the criteria one should use while deciding the threshold value. Explain the reason behind your choice of threshold in your model

```python
import logreg
import numpy as np
rec=[]
pre=[]
thre=np.arange(0.1,0.9,0.05)
for i in np.arange(0.1,0.9,0.05):
    p,r=logreg.run2(i)
    if(len(p)==2):
        #a=(p[0]+p[1])/2
        a=p[1]
        pre.append(a)
    else:
        pre.append(p)
    if(len(r)==2):
        #b=(r[0]+r[1])/2
        b=r[1]
        rec.append(b)
    else:
        rec.append(r)
        
    
import matplotlib.pyplot as plt
plt.plot(thre,pre,label="Precision")
plt.plot(thre,rec,label="Recall")
plt.xlabel("Threshold Value")
plt.ylabel("Precision-Recall")
plt.legend(loc="best")
plt.title("Precision Recall for Positive samples")
plt.grid(True)
plt.show()
```
